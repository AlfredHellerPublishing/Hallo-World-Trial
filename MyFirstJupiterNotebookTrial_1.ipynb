Her laver jeg min første grafik i Jupiter notebooken. Du kan run det ved at klicke "Shift-Enter" eller trykke på venstre side knappen, eller i menuen sige "run all".

```
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(20)
y = [x_i + np.random.randn(1) for x_i in x]
a, b = np.polyfit(x, y, 1)
_ = plt.plot(x, y, 'o', np.arange(20), a*np.arange(20)+b, '-')
```
